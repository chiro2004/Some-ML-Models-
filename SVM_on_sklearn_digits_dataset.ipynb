```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
```

```python
digitsdata=load_digits()
```

```python
dir(digitsdata)
```

```python
digitsdata.images[0]
```

```python
df=pd.DataFrame(digitsdata.data,columns=digitsdata['feature_names'])
```

```python
df.head()
```

```python
df['target']=digitsdata.target
```

```python
df.head()
```

```python
X=df.drop(['target'],axis='columns')
```

```python
y=df.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
svmodel=SVC(kernel='linear')
```

```python
svmodel.fit(X_train,y_train)
```

```python
svmodel.predict(X_test)
```

```python
y_test
```

```python
svmodel.score(X_test,y_test)
```
