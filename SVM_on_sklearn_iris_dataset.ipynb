```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
irisdata=load_iris()
```

```python
irisdata.feature_names
```

```python
df=pd.DataFrame(irisdata.data,columns=irisdata.feature_names)
```

```python
df.head()
```

```python
dir(irisdata)
```

```python
df['target']=irisdata.target
```

```python
df['flower_name']=df.target.apply(lambda x: irisdata.target_names[x])
```

```python
df.head()
```

```python
df0=df[df.target==0]
```

```python
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
X=df.drop(['target','flower_name'],axis='columns')
```

```python
y=df.target
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
svmodel=SVC(C=10,gamma=10,kernel='linear')
```

```python
svmodel.fit(X_train,y_train)
```

```python
svmodel.predict(X_test)
```

```python
svmodel.score(X_test,y_test)
```
